# Penggabungan Data Menggunakan Inner Join, Left Join, Right Join, dan Outer Join Dan Jawablah Pertanyaan

```python
import pandas as pd

# Baca data dari file CSV
tumor_size_ganjil = pd.read_csv('df1tumor_size_odd.csv')
tumor_size_genap = pd.read_csv('df2tumor_size_even.csv')
breast_quad_ganjil = pd.read_csv('df1breastquad_odd.csv')
breast_quad_genap = pd.read_csv('df2breastquad_even.csv')

# Inner Join
inner_join_ganjil = pd.merge(tumor_size_ganjil, breast_quad_ganjil, on='PatientID', how='inner')
print("Hasil Inner Join untuk ganjil:")
print(inner_join_ganjil)
inner_join_genap = pd.merge(tumor_size_genap, breast_quad_genap, on='PatientID', how='inner')
print("Hasil Inner Join untuk genap:")
print(inner_join_genap)

# Left Join
left_join_ganjil = pd.merge(tumor_size_ganjil, breast_quad_ganjil, on='PatientID', how='left')
print("Hasil Left Join untuk ganjil:")
print(left_join_ganjil)
left_join_genap = pd.merge(tumor_size_genap, breast_quad_genap, on='PatientID', how='left')
print("Hasil Left Join untuk genap:")
print(left_join_genap)

# Right Join
right_join_ganjil = pd.merge(tumor_size_ganjil, breast_quad_ganjil, on='PatientID', how='right')
print("Hasil Right Join untuk ganjil:")
print(right_join_ganjil)
right_join_genap = pd.merge(tumor_size_genap, breast_quad_genap, on='PatientID', how='right')
print("Hasil Right Join untuk genap:")
print(right_join_genap)

# Outer Join
outer_join_ganjil = pd.merge(tumor_size_ganjil, breast_quad_ganjil, on='PatientID', how='outer')
print("Hasil Outer Join untuk ganjil:")
print(outer_join_ganjil)
outer_join_genap = pd.merge(tumor_size_genap, breast_quad_genap, on='PatientID', how='outer')
print("Hasil Outer Join untuk genap:")
print(outer_join_genap)
```

```python
# Jumlah rekord dengan nilai atribut ukuran tumor dan kuadran kanker payudara
jumlah_record_lengkap_ganjil = inner_join_ganjil.shape[0]
print("Jumlah record pasien yang memiliki nilai atribut ukuran tumor dan kuadran kanker payudara (Ganjil):" , jumlah_record_lengkap_ganjil, "\n")
jumlah_record_lengkap_genap = inner_join_genap.shape[0]
print("Jumlah record pasien yang memiliki nilai atribut ukuran tumor dan kuadran kanker payudara (Genap):", jumlah_record_lengkap_genap, "\n")
total_semuanya = jumlah_record_lengkap_ganjil + jumlah_record_lengkap_genap
print("Jumlah record semuanya:", total_semuanya, "\n")
```

Lakukan analisis deskriptif berapa banyak atribut yang memiliki missing values 

```python
import pandas as pd

# Baca dataset hepatitis
data_hepatitis = pd.read_csv('Hepatitis.csv', header=1, index_col=0)

def ganti_pertanyaan_tanda_dengan_tidak_ada(a):
    if a == '?':
        return None
    else:
        return a

# Buat salinan data asli
update_data_hepatitis = data_hepatitis.copy()

# Terapkan fungsi penggantian pada setiap kolom
for kolom in update_data_hepatitis.columns:
    update_data_hepatitis[kolom] = update_data_hepatitis[kolom].apply(ganti_pertanyaan_tanda_dengan_tidak_ada)

# Informasi tentang dataset
info_data_hepatitis = update_data_hepatitis.info()
print("\nInformasi dataset hepatitis adalah :", info_data_hepatitis)

# Jumlah nilai yang hilang (NaN) dalam setiap kolom
jumlah_null_per_kolom = update_data_hepatitis.isna().sum()
print("\nJumlah nilai yang hilang dalam setiap kolom adalah :")
print(jumlah_null_per_kolom)
```

Lakukan imputasi missing values dengan berbagai teknik yang efektif dan baik untuk mengisi atribut-atribut yang NaN, misalnya menggunakan mean, median, modus, clustering, regression, maupun metode taksiran dan prediksi lainnya

```python
# memuat library
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
# Baca dataset hepatitis
data_hepatitis = pd.read_csv('Hepatitis.csv', header=1, index_col=0)

# menggantikan '?' dengan NaN
data_hepatitis.replace('?', np.nan, inplace=True)
data_hepatitis.dropna(inplace=True)

imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')
imputer_modus = SimpleImputer(strategy='most_frequent')

# Kolom-kolom numerik
numeric_cols = ['Age', 'Bilirubin', 'Alk Phosphate', 'SGOT', 'Albumin', 'Protime']

# mean
data_hepatitis[numeric_cols] = imputer_mean.fit_transform(data_hepatitis[numeric_cols])
print('Hasil data hepatitis mean\n', data_hepatitis[numeric_cols])

# median
data_hepatitis[numeric_cols] = imputer_median.fit_transform(data_hepatitis[numeric_cols])
print('\nHasil data hepatitis median\n', data_hepatitis[numeric_cols])

# modus
data_hepatitis[numeric_cols] = imputer_modus.fit_transform(data_hepatitis[numeric_cols])
print('\nHasil data hepatitis modus\n', data_hepatitis[numeric_cols])

# Imputasi dengan clustering, anda perlu memilih atribut yang sesuai untuk clustering
clustering_data_hepatitis = data_hepatitis[['Age', 'Bilirubin']]
kmeans = KMeans(n_clusters=3, random_state=0)
data_hepatitis['Cluster_Labels'] = kmeans.fit_predict(clustering_data_hepatitis)
print('Hasil clustering\n', data_hepatitis['Cluster_Labels'])
data_hepatitis['Alk Phosphate'] = data_hepatitis.groupby('Cluster_Labels')['Alk Phosphate'].transform(lambda x: x.fillna(x.mean()))
print('\nHasil kolom Alk Phosphate yang sudah diimputasi\n', data_hepatitis['Alk Phosphate'])

# Imputasi dengan regresi, anda perlu mengganti kolom target dan atribut yang digunakan untuk regresi sesuai kebutuhan Anda
regression_data_hepatitis = data_hepatitis[['Age', 'Bilirubin', 'Albumin']]
target_column = 'Alk Phosphate'
regressor = LinearRegression()
for target_value in data_hepatitis[target_column].unique():
    subset = data_hepatitis[data_hepatitis[target_column] == target_value]
    subset_X = subset[['Age', 'Bilirubin', 'Albumin']].dropna()
    subset_y = subset[subset[target_column].notna()][target_column]
    regressor.fit(subset_X, subset_y)
    mask = data_hepatitis[target_column] == target_value
    data_hepatitis.loc[mask, target_column] = regressor.predict(data_hepatitis[mask][['Age', 'Bilirubin', 'Albumin']])
print('Hasil dari target value data hepatitis adalah :', target_value)
print('Hasil dari regresi data hepatitis yang diambil kolom tertentu adalah :\n', data_hepatitis.loc[mask, target_column])
```

# Deteksi Outlier: InterQuartile Range dan BoxPlot 

Bank mana yang teramai dan tersepi bagi nasabah untuk melakukan penarikan uang? 

```python
# hitung transaksi setiap bank
gabungan_data_bank = data_bank.groupby('ATM Name')[['Total amount Withdrawn (Rupee)', 'Total amount Withdrawn (Rupiah)']].sum()
print('Hasil gabungan data bank\n', gabungan_data_bank) 

# tampilkan bank teramai
transaksi_bank_teramai = gabungan_data_bank[['Total amount Withdrawn (Rupee)', 'Total amount Withdrawn (Rupiah)']].idxmax()
print('\nBank mana yang teramai bagi nasabah untuk melakukan penarikan uang adalah :')
print(transaksi_bank_teramai)

# tampilkan bank tersepi
transaksi_bank_tersepi = gabungan_data_bank[['Total amount Withdrawn (Rupee)', 'Total amount Withdrawn (Rupiah)']].idxmin()
print('Bank mana yang tersepi bagi nasabah untuk melakukan penarikan uang adalah :')
print(transaksi_bank_tersepi)
```

Pada hari apa ATM paling sering dan pada hari apa yang paling jarang dikunjungi oleh nasabah untuk melakukan penarikan uang? 

```python
# memilih kolom numerik
kolom_numerik = data_bank.select_dtypes(include = [np.number])

# hitung jumlah transaksi penarikan uang setiap hari
data_bank_hari_kerja = kolom_numerik.groupby(data_bank['Weekday']).sum()
print('Hasil dari transaksi penarikan uang setiap hari\n',data_bank_hari_kerja)

# tampilkan hari yang sering dikunjungi
data_bank_sering_dikunjungi = data_bank['Weekday'].value_counts().idxmax()
print('\nHasil hari yang sering dikunjungi oleh nasabah adalah :',data_bank_sering_dikunjungi)

# tampilkan hari yang jarang dikunjungi
data_bank_jarang_dikunjungi = data_bank['Weekday'].value_counts().idxmin()
print('Hasil hari yang jarang dikunjungi oleh nasabah adalah :', data_bank_jarang_dikunjungi)
```

Lakukan deteksi outlier untuk melihat potensi adanya fraud berdasarkan atribut Total amount Withdrawn (in Rupiah) pada dataset ATM Transaction menggunakan metode InterQuartileRange dan visualisasikan menggunakan BoxPlot serta sebutkan nama No Of Withdrawals dan nama ATM-nya yang terdeteksi fraud. 

```python
# Hitung IQR Total amount Withdrawn (Rupiah)
Q1 = data_bank['Total amount Withdrawn (Rupiah)'].quantile(0.25)
Q3 = data_bank['Total amount Withdrawn (Rupiah)'].quantile(0.75)
IQR = Q3 - Q1
print('Hasil dari IQR total amount Withdrawn (Rupiah) adalah :', IQR)

# Hitung batas atas dan bawah
batas_atas = data_bank['Total amount Withdrawn (Rupiah)'].quantile(0.75) + 1.5 * IQR
print('Hasil dari batas atas Total amount Withdrawn (Rupiah) adalah :', batas_atas)
batas_bawah = data_bank['Total amount Withdrawn (Rupiah)'].quantile(0.25) - 1.5 * IQR
print('Hasil dari batas atas Total amount Withdrawn (Rupiah) adalah :', batas_bawah)

# outlier 
outlier_penarikan_uang_rupiah = data_bank[(data_bank['Total amount Withdrawn (Rupiah)'] < batas_bawah) 
                                | (data_bank['Total amount Withdrawn (Rupiah)'] > batas_atas)]
print('\nHasil dari outlier Total amount Withdrawn (Rupiah) adalah :\n', outlier_penarikan_uang_rupiah)

# boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data_bank['Total amount Withdrawn (Rupiah)'], vert = False)
plt.title('Box Plot dari Total amount Withdrawn (Rupiah)')
plt.show()

#sebutkan nama No Of Withdrawals dan nama ATM-nya yang terdeteksi fraud.
print('Data yang terdeteksi sebagai outlier :\n', outlier_penarikan_uang_rupiah[['No Of Withdrawals', 'ATM Name']])
```
