# Time Series Analysis

```python
#visualisasikan kasus baru covid19 dari beberapa tahun
import matplotlib.pyplot as plt
plt.figure(figsize = (12, 6))
plt.plot(data_covid19.index, data_covid19["New Cases"], marker = ".", linestyle = "-", color = "Green")
plt.title("Visualisasi Kasus Baru Covid19 dari Beberapa Tahun")
plt.xlabel("Waktu Kejadian")
plt.ylabel("Kasus Baru")
plt.grid(True)
plt.show()
```

```python
#visualisasikan kasus baru covid19 dari beberapa tahun
import matplotlib.pyplot as plt
plt.figure(figsize = (12,6))
plt.plot(kasus_pertahun, marker = ".", linestyle = "-", color = "Blue")
plt.title("Visualisasi Kasus Baru Covid19 dari Beberapa Tahun")
plt.xlabel("Waktu Kejadian")
plt.ylabel("Kasus Baru")
plt.grid(True)
plt.show()
```

```python
#grouping data time series dari beberapa bulan
gabungan_bulan = data_covid19.groupby(["Tahun", "Bulan"])
kasus_perbulan = gabungan_bulan["New Cases"].sum()
print(kasus_perbulan)
```

```python
#grouping data time series dari beberapa hari
gabungan_hari = data_covid19.groupby(["Tahun", "Bulan", "Hari"])
kasus_perhari = gabungan_hari["New Cases"].sum()
print(kasus_perhari)
```

```python
#visualisasikan kasus baru covid19 dari beberapa bulan
import matplotlib.pyplot as plt
tahun = kasus_perhari.index.get_level_values("Tahun")
bulan = kasus_perhari.index.get_level_values("Bulan")
hari = kasus_perhari.index.get_level_values("Hari")
gabungan_tahun_bulan_hari = [str(Tahun) + "-" + "{:02}".format(Bulan) + "-" + "{:02}".format(Hari) for Tahun, Bulan, Hari in zip(tahun, bulan, hari)]
plt.figure(figsize = (12, 6))
plt.plot(gabungan_tahun_bulan_hari, kasus_perhari, marker = ".", linestyle = "-", color = "Red")
plt.title("Visualisasi Kasus Baru Covid19 dari Beberapa Hari")
plt.xlabel("Tahun Bulan Hari")
plt.ylabel("Kasus Baru")
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

# Hypothesis Testing

```python
from scipy.stats import ttest_1samp
import numpy as np
data_yang_diambil = kasus_perhari
#rata data
rata_data = np.mean(data_yang_diambil)
h0 = rata_data
h1 = rata_data
#uji t
t_statistik, p_value = ttest_1samp(data_yang_diambil, h0, alternative = 'greater')
alpha = 0.05
#uji hipotesis
print("H0 adalah hipotesis nol")
print("H1 adalah hipotesis alternatif")
print("Hasil dari rata-rata data :", rata_data)
print("H0 : rata-rata dari kasus baru harian sama dengan", rata_data)
print("H1 : rata-rata dari kasus baru harian lebih dari", rata_data)
print("Hasil dari t-statistik :", t_statistik)
print("Hasil dari p-value :", p_value)
print("Nilai alpha :", alpha)
#kesimpulan
if p_value < alpha :
    print("Hipotesis nol ditolak karena p-value ({}) < alpha ({})".format(p_value, alpha))
    print("Oleh karena itu kita bisa simpulkan bahwa rata-rata kasus baru harian lebih dari", rata_data)
else :
    print("Hipotesis nol gagal tolak karena p-value ({}) >= alpha ({})".format(p_value, alpha))
    print("Oleh karena itu kita bisa simpulkan bahwa rata-rata kasus baru harian sama dengan", rata_data)
```

# Analisis Korelasi

```python
def korelasi_pearson(x,y):
    n_atau_total_data = len(x)
    jumlah_x = sum(x)
    jumlah_y = sum(y)
    jumlah_xkuadrat = sum(x**2 for x in x)
    jumlah_ykuadrat = sum(y**2 for y in y)
    jumlah_xy = sum(x*y for x, y in zip(x,y))
    pembilang = (n_atau_total_data*jumlah_xy)-(jumlah_x*jumlah_y)
    penyebut = ((n_atau_total_data*jumlah_xkuadrat - jumlah_x**2)*(n_atau_total_data*jumlah_ykuadrat-jumlah_y**2))**0.5
    korelasi = pembilang/penyebut
    return korelasi
```

```python
variabel_x = "New Cases"
variabel_y = "New Deaths"
x = data_covid19[variabel_x]
y = data_covid19[variabel_y]
korelasi = korelasi_pearson(x,y)
print("Hasil nilai korelasi pearson dari variabel ({}) dan ({}) : {}".format(variabel_x, variabel_y, korelasi))
```

# Regresi Dan Evaluasi

```python
import numpy as np

def regresi_linier_sederhana(x,y):
    n_atau_total_data = len(x)
    rata2_x = np.mean(x)
    rata2_y = np.mean(y)
    sxx = sum((x-rata2_x)**2 for x in x)
    sxy = sum((x-rata2_x)*y for x, y in zip(x,y))
    b1_topi = sxy/sxx
    b0_topi = rata2_y - (b1_topi*rata2_x)
    ytopi = [b0_topi+(b1_topi*x) for x in x]
    ssr = sum((ytopi-rata2_y)**2 for ytopi in ytopi)
    sse = sum((y-ytopi)**2 for y, ytopi in zip(y, ytopi))
    sst = sum((y-rata2_y)**2 for y in y)
    msr = ssr/1
    mse = sse/(n_atau_total_data-2)
    rsquared = ssr/sst
    fhit = msr/mse
    thit = b1_topi/((mse/sxx)**0.5)
    return b1_topi, b0_topi, rsquared, fhit, thit, ytopi
def kalkulator_mape(aktual, prediksi):
    nilai_mape = sum(abs((aktual-prediksi)/aktual))/len(aktual)*100
    return nilai_mape
```

```python
variabel_x = "New Cases"
variabel_y = "New Deaths"
x = data_covid19[variabel_x]
y = data_covid19[variabel_y]
b1_topi, b0_topi, rsquared, fhit, thit, ytopi = regresi_linier_sederhana(x,y)
nilai_mape = kalkulator_mape(y, ytopi)
print("Hasil nilai b1 topi :", b1_topi)
print("Hasil nilai b0 topi :", b0_topi)
print("Hasil model regresi linier sederhana ytopi = {} + {}x".format(b0_topi, b1_topi))
print("Hasil nilai rkuadrat :", rsquared)
print("Hasil nilai fhit :", fhit)
print("Hasil nilai thit :", thit)
print("Evaluasi model regresi dengan menggunakan MAPE antara variabel ({}) dan ({})".format(variabel_x, variabel_y))
print("Hasil nilai MAPE :", nilai_mape)
```
