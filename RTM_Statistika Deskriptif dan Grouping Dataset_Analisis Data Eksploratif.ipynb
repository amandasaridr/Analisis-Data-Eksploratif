# Tabel Distribusi Uniform

Carilah nilai a dan b terlebih dahulu

```python
import math

# memuat data ph
data_ph = [[3.5], [1.5], [8.2], [2.6], [4.5],
          [9.1], [2.7], [6.8], [4.3], [3.5],
          [3.6], [7.4], [7.3], [8.9], [7.1],
          [2.1], [9.5], [2.8], [2.3], [8.4],
          [8.6], [12], [2.1], [3.5], [1.5],
          [6.7], [5.5], [6.7], [2.6], [5.5],
          [6.6], [6.9], [8.3], [8.9], [7.5],
          [4.3], [9.7], [2.6], [3.3], [8.8]]

# menggunakan nilai a dan nilai b untuk mendapatkan nilai minimum dan maksimum
a = min(min(baris) for baris in data_ph)
b = max(max(baris) for baris in data_ph)
print("Nilai yang paling kecil adalah :", a)
print("Nilai yang paling besar adalah :", b)
```

Gunakan persamaan mean dan standard deviation sebagai bekal untuk mendapatkan nilai a dan b sesuai dengan sifat dasar statistika dalam distribusi uniform berikut 

```python
# menghitung nilai a dan b untuk mencari mean dan standard deviation
mean = (a + b) / 2
std_dev = (b-a) / math.sqrt(12)
print("Hasil dari mean data ph adalah :", mean)
print("Hasil dari standart deviation adalah :", std_dev)
```

Berapa persentase probabilitas x < 5 berdasarkan persamaan dalam distribusi uniform

```python
# menghitung persentase probalitas x < 5
x = 5
probabilitas = (x - a) / (x - b)
persentase = probabilitas * 100
print("Hasil dari probabilitas data ph adalah :", probabilitas)
print("Hasil dari persentasi probalitas data ph adalah :", persentase, "%")
```

Berapa probabilitas distribusi uniform dari nilai a dan b

```python
from scipy.integrate import quad

# menghitung probabilitas distribusi uniform
def probabilitas_distribusi_uniform(x1, x2):
    return quad(lambda x: 1 / (b - a), x1, x2)[0]

a = 1.5
b = 12

x1 = 0
x2 = 5

prob = probabilitas_distribusi_uniform(x1, x2)
print("Hasil dari probabilitas distribusi uniform adalah :", prob)
```

# Tabel Distribusi Normal

```python
import pandas as pd
import numpy as np

# membaca data dari csv
data_normal_distribusi = pd.read_csv("normal_distribution.csv")
data_normal_distribusi
```

Hitung persamaan f(x) dari nilai x data tersebut dengan menggunakan persamaan distribusi normal

```python
# data distribusi normal dengan format yang salah
data_normal_distribusi = pd.DataFrame({'x': ['-5,84', '-4,86', '-3,96', '-3,74', '-2,86', '-2,35', '-2,25', '-2,16', '-1,8', '-1,69', '-1,53', '-1,52', '-1,17', '-1,17', '-1,12', '0,94', '0,89', '0,86', '0,73', '0,69', '0,44', '0,37', '0,28', '0,20', '0,15', '0,90', '0,12', '0,14', '0,31', '0,33', '0,40', '0,56', '0,73', '0,73', '0,91', '0,92', '1,20', '1,21', '1,55', '1,80', '1,90', '1,91', '1,91', '1,93', '2,14', '2,45', '3,00', '3,13', '3,54', '4,12', '4,85', '5,96']})

# mengganti koma sebagai pemisah desimal dengan titik
data_normal_distribusi['x'] = data_normal_distribusi['x'].str.replace(',', '.')

# mengonversi kolom 'x' menjadi float
data_normal_distribusi['x'] = data_normal_distribusi['x'].astype(float)

# menghitung mu dan sigma
mu = np.mean(data_normal_distribusi['x'])
sigma = np.std(data_normal_distribusi['x'])
print("Rata-rata (mu) dari data distribusi normal adalah:", mu)
print("Deviasi Standar (sigma) dari data distribusi normal adalah:", sigma)

# menghitung f(x) pada data normal distribusinya
data_normal_distribusi['f(x)'] = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(data_normal_distribusi['x'] - mu)**2 / (2 * sigma**2))
data_normal_distribusi
```

Visualisasikan antara nilai x dan f(x)

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# menentukan data berdistribusi normal atau tidak
statistic, p_value = shapiro(data_normal_distribusi['x'])
alpha = 0.05
if p_value > alpha:
    print("Data ini berdistribusi normal")
else :
    print("Data ini tidak berdistribusi normal")

# visualisasi data normal distribusi
plt.plot(data_normal_distribusi['x'], data_normal_distribusi['f(x)'])
plt.title("Visualisasi Data Normal DIstribusi")
plt.xlabel("Nilai x")
plt.ylabel("Nilai f(x)")
plt.grid(True)
plt.show()
```

# Tabel Skewness

```python
import pandas as pd
import numpy as np

data_skewness = pd.read_csv("skewness.csv")
data_skewness
```

Hitunglah persamaan f(x) dari nilai x dari persamaan distribusi normal

```python
# mengambil kolom x saja
kolom_x_saja = data_skewness['x']

# mengganti koma ke titik
kolom_x_saja_titik = kolom_x_saja.str.replace(',','.')

# kolom x menjadi float
kolom_x_float = kolom_x_saja_titik.astype(float)

# menghitung mu dan sigma
mu = np.mean(kolom_x_float)
sigma = np.std(kolom_x_float)
print("Rata-rata (mu) dari data distribusi normal adalah:", mu)
print("Deviasi Standar (sigma) dari data distribusi normal adalah:", sigma)

# menghitung f(x) pada data normal distribusinya
data_skewness['f(x)'] = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(kolom_x_float - mu)**2 / (2 * sigma**2))
data_skewness
```

Visualisasikan antara nilai x dan f(x) dalam diagram garis 

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# menentukan data berdistribusi normal atau tidak
statistic, p_value = shapiro(kolom_x_float)
alpha = 0.05
if p_value > alpha:
    print("Data ini berdistribusi normal")
else :
    print("Data ini tidak berdistribusi normal")

# visualisasi data normal distribusi
plt.plot(kolom_x_float, data_skewness['f(x)'])
plt.title("Visualisasi Data Normal DIstribusi")
plt.xlabel("Nilai x")
plt.ylabel("Nilai f(x)")
plt.grid(True)
plt.show()
```

Gunakan persamaan skewness

```python
# menghitung mean
mean = np.mean(kolom_x_float)
print("Hasil mean dari data ini adalah :", mean)

# menghitung median
median = np.median(kolom_x_float)
print("Hasil median dari data ini adalah :", median)

# menghitung standard deviasi
std_dev = np.std(kolom_x_float)
print("Hasil standart deviasi data ini adalah :", std_dev)

# menghitung skewness
skewness = 3*(mean - median) / std_dev
print("Hasil dari skewness data ini adalah :", skewness)

# menentukan arah condong distribusi data ini
if skewness > 0 :
    print("Distribusi data yang condong kearah positif")
    print("Nilai skewness", skewness)
    
elif skewness < 0 :
    print("Distribusi data yang condong kearah negatif")
    print("Nilai skewness", skewness)
    
else : 
    print("Distribusi data yang normal")
```

# Tabel Kurtosis

```python
import pandas as pd
import numpy as np

data_kurtosis = pd.read_csv("kurtosis.csv")
data_kurtosis
```

Hitung persamaan f(x) dari nilai x 

```python
# mengambil kolom x
kolom_x = data_kurtosis['x']

# mengganti koma ke titik
kolom_x_titik = kolom_x.str.replace(',','.')

# kolom x menjadi float
kolom_x_float = kolom_x_titik.astype(float)

# menghitung mu dan sigma
mu = np.mean(kolom_x_float)
sigma = np.std(kolom_x_float)
print("Rata-rata (mu) dari data distribusi normal adalah:", mu)
print("Standart deviasi (sigma) dari data distribusi normal adalah:", sigma)

# menghitung f(x) pada data normal distribusinya
data_kurtosis['f(x)'] = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(kolom_x_float - mu)**2 / (2 * sigma**2))
data_kurtosis
```

Visualisasikan antara nilai x dan f(x)

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# menentukan data berdistribusi normal atau tidak
statistic, p_value = shapiro(kolom_x_float)
alpha = 0.05
if p_value > alpha:
    print("Data ini berdistribusi normal")
else :
    print("Data ini tidak berdistribusi normal")

# visualisasi data normal distribusi
plt.plot(kolom_x_float, data_kurtosis['f(x)'])
plt.title("Visualisasi Data Normal DIstribusi")
plt.xlabel("Nilai x")
plt.ylabel("Nilai f(x)")
plt.grid(True)
plt.show()
```

Gunakan persamaan kurtosis 

```python
# menghitung xbar dan s4nya
banyak_data_kurtosis = len(kolom_x_float)
print("Hasil dari banyaknya data kurtosis adalah :", banyak_data_kurtosis)
x_bar = sum(kolom_x_float) / banyak_data_kurtosis
print("Hasil dari x bar data kurtosis adalah :", x_bar)
s_pangkat_empat = np.std(kolom_x_float)
n = 43

# menghitung kurtosisnya
alfa_empat = (1/n) * np.sum((kolom_x_float - x_bar) ** 4) / (s_pangkat_empat**4)
print("Hasil data kurtosis ini adalah :", alfa_empat)

# menentukan data mengarah platykurtic, mesokurtic, atau leptokurtic
if alfa_empat > 3 :
    print("Leptokurtic", "Hasil nilai kurtosis adalah :", alfa_empat)
elif alfa_empat < 3 :
    print("Platykurtic", "Hasil nilai kurtosis adalah :", alfa_empat)
else :
    print("Mesokurtic", "Hasil nilai kurtosis adalah :", alfa_empat)
```

# Grouping Dataset Pivoting Table

```python
import pandas as pd

data_bahan_pokok = pd.read_csv("data_bahan_pokok.csv")
data_bahan_pokok
```

Pivoting tabel berdasarkan jumlah pembelian bahan pokok (kg) per kuatil

```python
# tabel pivot
data_bahan_pokok_pivot = data_bahan_pokok.pivot_table(values = 'JUMLAH', index = ['BARANG', 'KUARTAL'], columns = 'TAHUN', aggfunc = sum, fill_value = 0)

# memasukkan kolom grand total
data_bahan_pokok_pivot['GRAND TOTAL'] = data_bahan_pokok_pivot.sum(axis = 1)
data_bahan_pokok_pivot
```

Pivoting Tabel Berdasarkan Total Pembelian Bahan Pokok (Rupiah) per Kuartal

```python
# tabel pivot
data_bahan_pokok_pivot = data_bahan_pokok.pivot_table(values = 'TOTAL', index = ['BARANG', 'KUARTAL'], columns = 'TAHUN', aggfunc = sum)

data_bahan_pokok_pivot['GRAND TOTAL'] = data_bahan_pokok_pivot.sum(axis = 1)
data_bahan_pokok_pivot
```
