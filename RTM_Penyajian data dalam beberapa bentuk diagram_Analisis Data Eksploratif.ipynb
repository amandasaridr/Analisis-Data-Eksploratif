# Data Daftar Dosen UPN Veteran X Angkatan 2022

```python
import pandas as pd
import matplotlib.pyplot as plt

#membaca data dari file csv yang sudah dibuat
data_pertama = pd.read_csv("daftar_dosen_upn_veteran_x_angkatan_2022.csv")

#menambah kolom gelar dan universitas kemudian hapus kolom No. dan Pendidikan terakhir
data_pertama["Gelar"] = data_pertama["Pendidikan Terakhir"].apply(lambda x: x.split(",")[0])
data_pertama["Universitas"] = data_pertama["Pendidikan Terakhir"].apply(lambda x: x.split(",")[-1].strip())
data_pertama.drop(columns=["No.", "Pendidikan Terakhir"], inplace=True)

#menampilkan hasil akhir datanya
print("\nHasil data baru yang baru\n")
print(data_pertama)
```

```python
#praproses kolom gelar untuk mengelompokkan jenis pendidikan
def jenis_gelar(baris):
    if "S1" in baris:
        return "S1"
    elif "S2" in baris:
        return "S2"
    elif "S3" in baris:
        return "S3"
    elif "Post-doctoral" in baris:
        return "Post-doctoral"
    else:
        return "Gelar Lainnya"
data_pertama["Jenis Pendidikan"] = data_pertama["Gelar"].apply(jenis_gelar)

#menghitung jumlah jenis pendidikan
jenis_pendidikan_dosen = data_pertama["Jenis Pendidikan"].value_counts()

#membuat diagram lingkaran
plt.figure(figsize = (8,8))
plt.pie(jenis_pendidikan_dosen, labels = jenis_pendidikan_dosen.index, autopct = "%1.1f%%", startangle = 140)
plt.title("Jumlah Pendidikan dari Jenis Gelar Dosen")
plt.show()
```

# Data Daftar Pasien Diabetes Berdasarkan Faktor Gula Darah, Berat Badan dan Tinggi Badan

```python
import pandas as pd
import matplotlib.pyplot as plt

#membaca data dari file CSV
data_kedua = pd.read_csv("daftar_pasien_diabetes_berdasarkan_faktor_gula_darah_berat_badan_dan_tinggi_badan.csv")

#menghapus kolom No.
data_kedua.drop("No.", axis=1, inplace=True)

#Menghitung BMI (Body Mass Index)
data_kedua["BMI"] = data_kedua["Berat Badan (kg)"] / ((data_kedua["Tinggi Badan (cm)"] / 100) ** 2)
print("\nHasil nilai kategori berat badan\n")
print(data_kedua["BMI"])

# Membuat fungsi untuk mengkategorikan berat badan
def kategori_berat_badan(bmi): 
    if bmi < 18.5: 
        return "Kurang berat badan" 
    elif 18.5 <= bmi < 24.9: 
        return "Normal"
    elif 25 <= bmi < 29.9: 
        return "Berat badan berlebih"
    else: 
        return "Berat badan sangat berlebih"

# Mengkategorikan berat badan
data_kedua["Kategori Berat Badan"] = data_kedua["BMI"].apply(kategori_berat_badan)
print("\nKategori dari hasil nilai tersebut\n")
print(data_kedua["Kategori Berat Badan"])

# Visualisasi data dengan scatter plot
plt.figure(figsize=(8, 6)) 

# Memisahkan data berdasarkan kategori berat badan
for category in data_kedua["Kategori Berat Badan"].unique():
    subset = data_kedua[data_kedua["Kategori Berat Badan"] == category]
    plt.scatter(subset["BMI"], subset["Gula Darah (Puasa)"], label=category)

plt.title("Hubungan antara BMI dan Gula Darah") 
plt.xlabel("BMI") 
plt.ylabel("Gula Darah")

# Menambahkan legenda
plt.legend()
plt.grid(True)
plt.show()
```

# Data Daftar Kasus Pasien Terkonfirmasi Covid-19 Harian

```python
import pandas as pd
import matplotlib.pyplot as plt

#membaca data dari file CSV
data_ketiga = pd.read_csv("daftar_kasus_pasien_terkonfirmasi_covid-19_harian.csv")

#menghapus kolom No
data_ketiga.drop("No.", axis=1, inplace=True)

#menghitung pertumbuhan kasus covid harian
data_ketiga["Pertumbuhan kasus covid harian"] = data_ketiga["Jumlah Kasus Terkonfirmasi"].diff()
data_ketiga["Pertumbuhan kasus covid harian"].fillna(0, inplace=True)
print("\nHasil dari pertumbuhan kasus covid harian\n")
print(data_ketiga["Pertumbuhan kasus covid harian"])

# Membuat diagram
plt.figure(figsize=(12, 6))
plt.plot(data_ketiga['Tanggal'], data_ketiga['Pertumbuhan kasus covid harian'], marker='o', linestyle='-', color='blue')
plt.plot(data_ketiga['Tanggal'], data_ketiga['Jumlah Kasus Terkonfirmasi'], marker='o', linestyle='-', color='red')
plt.title("Pertumbuhan Kasus COVID-19 Harian")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah yang terkena")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
