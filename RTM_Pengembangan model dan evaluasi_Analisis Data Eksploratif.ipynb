# Input data

```python
import pandas as pd

data_hewan = pd.read_csv("animal_morphology_classify.csv")
data_hewan
```

# Menghitung frekuensi jumlah data dan probabilitas pada tiap kelas/label 

```python
total = len(data_hewan)
insekta = len(data_hewan[data_hewan["Kelas"] == "Insekta"])
mamalia = len(data_hewan[data_hewan["Kelas"] == "Mamalia"])
hasil_probabilitas_mamalia = mamalia / total
hasil_probabilitas_insekta = insekta / total
print("Hasil probabilitas dari kelas mamalia :",hasil_probabilitas_mamalia)
print("Hasil probabilitas dari kelas insekta :",hasil_probabilitas_insekta)
```

# Menghitung frekuensi jumlah data dan probabilitas tiap fitur pada masing-masing kelas

```python
def hitung_probabilitas_fitur_per_kelas(fitur, nilai, kelas_yang_diberikan):
    subkelas = data_hewan[data_hewan["Kelas"] == kelas_yang_diberikan]
    hitung_fitur_kelas = len(subkelas[subkelas[fitur] == nilai])
    hitung_kelas = len(subkelas)
    faktor = 1
    probabilitas_fitur_kelas = (hitung_fitur_kelas + faktor) / (hitung_kelas + faktor * len(data_hewan))
    return probabilitas_fitur_kelas
```

# Kalikan semua probabilitas data pada semua fitur dan probabilitas kelas

```python
def prediksi_untuk_kelas(nama_hewan_yang_diinput):
    try:
        instansi_baru = data_hewan[data_hewan["Indeks"] == nama_hewan_yang_diinput].iloc[0]
        hasil_probabilitas_insekta_baru = hasil_probabilitas_insekta
        hasil_probabilitas_mamalia_baru = hasil_probabilitas_mamalia
        for fitur in data_hewan.columns[2:-1]:
            nilai_yang_baru = instansi_baru[fitur]
        hasil_probabilitas_insekta_baru *= hitung_probabilitas_fitur_per_kelas(fitur, nilai_yang_baru, "Insekta")
        hasil_probabilitas_mamalia_baru *= hitung_probabilitas_fitur_per_kelas(fitur, nilai_yang_baru, "Mamalia") 
        hasil_prediksi_kelas = "Mamalia" if hasil_probabilitas_mamalia_baru > hasil_probabilitas_insekta_baru else "Insekta"
        print(f"Hasil probabilitas insekta : {hasil_probabilitas_insekta_baru}")
        print(f"Hasil probabilitas mamalia : {hasil_probabilitas_mamalia_baru}")
        print(f"Hasil prediksi kelas dari {nama_hewan_yang_diinput} : {hasil_prediksi_kelas}")
    except IndexError: 
        print(f"Hewan dengan nama ini {nama_hewan_yang_diinput} tidak masuk kedalam dataset atau tidak temukan")
```

# Bandingkan hasil probabilitas antara kelas mamalia dan insekta

```python
nama_hewan_yang_diinput = input("Masukkan nama hewan yang ingin dicari hasilnya dan jenisnya :")
prediksi_untuk_kelas(nama_hewan_yang_diinput)
```
